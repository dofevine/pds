```python
import numpy as np
from matplotlib import pyplot as plt
from scipy.signal import freqz
```

questao 10

```python
banda_passante1 = 150
banda_passante2 = 250
faixa_transicao = 50
amostragem = 1000

# Freq admensionalizadas
fca1 = banda_passante1/amostragem
fca2 = banda_passante2/amostragem
dfa = faixa_transicao/amostragem

# Determinacao do tamanho da janela Hamming
N = round(5.5/dfa)
if not N % 2: N += 1
print(N)
n = np.arange(-N/2,N/2, dtype=int)
w = 0.42 + 0.5 * np.cos(2*np.pi*n/(N-1))+0.08*np.cos(4*np.pi*n/(N-1))

plt.stem(n, w, use_line_collection=True)
plt.title("Janela blackman")
plt.xlabel("n")
plt.ylabel("Amplitude (admensional)")
```

```python
window = np.blackman(N)
plt.stem(window)
plt.title("Janela blkackman")
```

resposta ao impulso ideal

```python
hd0 = np.array([2*(fca2-fca1)])
print(hd0)
wca1 = 2*np.pi*fca1
wca2 = 2*np.pi*fca2
nn = np.arange(1, N/2, dtype=int)

hd1 = 2*fca2*np.sin(nn*wca2)/(nn*wca2)-2*fca1*np.sin(nn*wca1)/(nn*wca1)
print(len(hd1))

# Lado esquerdo
hde = np.flip(hd1)

hd = np.concatenate((hde,hd0,hd1), axis = 0)
print(hd1[0])

plt.figure()

plt.stem(n,hd, use_line_collection=True)
print(hd.max())
```

```python
b = w * hd
plt.stem(n, b, use_line_collection=True)
plt.title("Filtro FIR")
plt.xlabel("n")
plt.ylabel("Amplitude (admensional)")
```

```python
w, h = freqz(b,1,512, amostragem)
magh = 20 * np.log10(np.abs(h))
fase = np.rad2deg(np.angle(h))

w = (amostragem * w) /(2*np.pi)

plt.figure()
plt.plot(w, magh)
plt.grid(True)
plt.figure()
plt.grid(True)
plt.plot(w, fase)
# plt.title("Filtro FIR")
# plt.xlabel("n")
# plt.ylabel("Amplitude (admensional)")
```
